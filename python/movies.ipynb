```python
headers = []
headers = movies_table[0].copy()

headers.extend(["tmdb_id", "overview", "poster_url", "release_date", "genre", "rating"])
print(headers)
print(len(headers))
```

```python
def urlify(in_string):
    return "%20".join(in_string.split())

def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)
```

```python
def add_ids_and_years(movie_row):
    title = movie_row[1]
    title =  ''.join(l for l in title if l not in string.punctuation)
    #print(title)
    movie_string = urlify(title)
    url = "https://api.themoviedb.org/3/search/movie?api_key=" + key + "&language=en-US&include_adult=false&query=" + movie_string

    print("/search/", url)
    
    film = requests.get(url)
    film = film.json()
    
    film_index = 0
    
    #For now, override to snag correct movie
    if "Dragon Blade" in title:
        film_index = 1
    elif "Master of Puppets" in title:
        return movie_row
    else: 
        film_index = 0
    
    #print("film_index:", film_index)
    film_overview = film['results'][film_index]['overview']
    film_poster_path = film['results'][film_index]['poster_path']
    film_id = film['results'][film_index]['id']
    film_year = film['results'][film_index]['release_date']
    film_year = datetime.strptime(film_year, "%Y-%m-%d").year
    
    movie_update = movie_row
    movie_update.append(film_id)
    movie_update.append(film_overview)
    movie_update.append(film_poster_path)
    movie_update.append(film_year)

    return movie_update
```

```python
def add_genres(movie_row):
    row = movie_row
    id_index = headers.index("tmdb_id")
    movie_id = str(movie_row[id_index])
    url = "https://api.themoviedb.org/3/movie/" + movie_id + "?api_key=" + key + "&language=en-US"
    print("/movie/ ", url)
    
    film = requests.get(url)
    film = film.json()
    genre = film["genres"][0]["name"].lower()
    rating = film["vote_average"]
    
    row.append(genre)
    row.append(rating)

    return row
```

```python
def create_obj_from_arr(arr):
    obj = {}
    for index, item in enumerate(arr):
        
        if headers[index] == "movie":
            obj["title"] = arr[index]
        elif headers[index] != "":
            obj[headers[index]] = item
    return obj
```
